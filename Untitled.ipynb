```python

import math
import matplotlib.pyplot as plt
import numpy as np


scale_factor = 25
z = 0
x_list = []
y_list = []

for theta in np.linspace(0, 10*math.pi, 200):
    r = scale_factor * math.cos(-0.4 * theta)
    x_list.append(r*np.cos(theta))
    y_list.append(r*np.sin(theta))

plt.plot(x_list,y_list)
plt.show()
```
