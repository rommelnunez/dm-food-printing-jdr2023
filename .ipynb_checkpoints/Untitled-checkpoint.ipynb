```python

import math
import matplotlib.pyplot as plt
import numpy as np


scale_factor = 5
z = 0
x_list = []
y_list = []

theta = np.radians(np.linspace(0,360*5,1000))
r = scale_factor * math.cos(-0.4 * theta)
x_2 = r*np.cos(theta)
y_2 = r*np.sin(theta)
plt.figure(figsize=[10,10])
plt.plot(x_2,y_2)
plt.show()


for i in range(0,108,1):  # draw 10 triangles
    j = i * .3
    r = scale_factor * math.cos(-.4 * j)
    # Calculate coordinates of the triangle vertices
    x = r * math.cos(j)
    y =r * math.sin(j)
    x_list.append(x)
    y_list.append(y)
    
plt.plot(x_list,y_list)  
plt.show()
```
